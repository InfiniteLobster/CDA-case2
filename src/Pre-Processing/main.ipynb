### Splitting

### 

```python
for col in sens_hr_df.columns:
    plt.figure(figsize=(6, 4))
    sens_hr_df.boxplot(column=col)
    plt.title(f"Boxplot of {col}")
    plt.xticks(rotation=90)
    plt.show()
```

```python
sens_hr_df_std = (sens_hr_df - sens_hr_df.mean()) / sens_hr_df.std()
sens_hr_df_std.plot.box(figsize=(8, 5))
plt.xticks(rotation=90)
plt.show()
```

```python
sens_temp_df_std = (sens_temp_df - sens_temp_df.mean()) / sens_temp_df.std()
sens_temp_df_std.plot.box(figsize=(8, 5))
plt.xticks(rotation=90)
plt.show()
```

```python
sens_eda_p_df_std = (sens_eda_p_df - sens_eda_p_df.mean()) / sens_eda_p_df.std()
sens_eda_p_df_std.plot.box(figsize=(8, 5))
plt.xticks(rotation=90)
plt.show()
```

```python
sens_eda_t_df_std = (sens_eda_t_df - sens_eda_t_df.mean()) / sens_eda_t_df.std()
sens_eda_t_df_std.plot.box(figsize=(8, 5))
plt.xticks(rotation=90)
plt.show()
```
